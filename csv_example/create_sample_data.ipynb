```python
import numpy as np
cov = [[1, 0.5], [0.5, 1]]
mean = (5, 10)
pts = np.random.multivariate_normal(mean, cov, size=1000)
```

```python
import matplotlib.pyplot as plt
plt.plot(pts[:, 0], pts[:, 1], '.', alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(pts)
df.columns = ['sample_A', 'sample_B']
df.index.name = 'gene'
df.to_csv("case.csv")
```
